### Forward Fill and Backward Fill

```python
disease.head()
```

```python
disease.fillna(method = 'ffill') # Missing value replaced with previous value
```

```python
disease.fillna(method="bfill") # Missing value replaced with next value
```

'ffill' and 'bfill' do not fill up the values if there is no value at the extremes. If there is a value at the center only then 'ffill' only replaces values in the latter half and opposite for 'bfill'

```python
n_disease = disease.fillna(method = "ffill")
n_disease = n_disease.fillna(method = "bfill")
n_disease.head()
```

### Restructuring Data

```python
weather = pd.read_csv("Data/weather.csv")
weather.head()
```

```python
weather.shape
```

```python
# 'melt' used to restructure into new column (creates two new columns- 1 for name of column and 1 for the values in the column)
# 'melt' multiple different columns into single one
new_weather_melt = pd.melt(weather,
        id_vars = ['year', 'month', 'element'],
        var_name = 'day',
        value_name = "temp"
       )
new_weather_melt.head(7)
```

```python
new_weather_pivot = new_weather_melt.pivot_table(index = ["year", "month", "day"],
                                                columns = "element",
                                                values = "temp", 
                                                dropna = False).reset_index()
new_weather_pivot.head(10)
```

```python
new_weather_pivot.to_csv('Data/new_weather.csv')
```

```python
new_weather_pivot.loc[(new_weather_pivot.month == 3) & (new_weather_pivot.day == "day2")]
```

```python
def mean_func(x):
    return sum(x) / len(x)
```

```python
new_weather_pivot[["min", "max"]].apply(mean_func, axis = 0)
```

### Merging Data Frames

```python
users = pd.DataFrame({'id':[1, 2, 3], 'name': ["A", "B", "C"]})
users
```

```python
msgs = pd.DataFrame({"userid": [1, 1, 2], "msg": ["hello", "hi", "bye"]})
msgs
```

```python
users.merge(msgs, left_on='id', right_on = 'userid')
```
